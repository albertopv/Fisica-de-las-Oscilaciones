# **CASO 1: Oscilación amortiguada en un sistema masa-resorte con rozamiento**

**Objetivo**

 * Determinar para qué valores de la masa ($m$), de la constante de elasticidad ($k$) y coeficiente de amortiguamiento ($b$) un sistema masa-resorte oscila de forma sub-amortiguada, critica y sobre-amortiguada.

 * Realizar la gráfica de la posición en funcion del tiempo $x(t)$.

**Modelo matematico**

El sistema masa-resorte con amortiguamiento se describe mediante la ecuación diferencial :

$m \frac{d^2x}{dt^2}+b\frac{dx}{dt}+kx = 0$   $\quad$  (1)

donde:

 * $m$ es la masa,
 * $b$ es el coeficiente de amortiguamiento,
 * $k$ es la constante elástica del resorte.

**Condiciones iniciales**:

Al sistema se le proporciona una determinada energía (cinética y/o potencial eleastica). Luego, en el instante en que se comienza a contar el tiempo, se le imprime una deformación $x_0$ y una velocidad $v_0$ para hacerlo oscilar. Es decir:

* $t=0$: $x=x_0$ y  $v=v_0$.   $\quad$  (2)

**Solución Genaral: Fórmula para calcular la posición en función del tiempo $x(t)$**

Se supone que: $m$, $b$ y $k$ no cambian duarante la oscilación, no existe fuerza externa al sistema durante las oscilaciones (La fuerza externa solo actua para hacerlo oscilar), se estudia la variación de la posición $x$ solamente respecto al tiempo ($t$). Es así, que la ED (1) es:

* ordinaria (solamente respecto al tiempo),
* lineal (la solución es una combinación lineal de soluciones)
* de segundo orden (necesito dos condiciones iniciales para hallar una solución particular),
* homogénea (Las oscilaciones son libres, no forzadas) y
* con coeficientes constantes (los parametos del sistema permanecen constantes)

La ED (1) se puede escribir así:

$\frac{d^2x}{dt^2} + 2 \alpha \frac{dx}{dt}+ \omega_0^2 x = 0$   $\quad$  (3)

donde:

* $\alpha = \frac{b}{2m}$: $\quad $ Es un factor de amortiguamiento; y
* $\omega_0 = \sqrt{\frac{k}{m}}$: $\quad$ es la frecuencia natural del sistema


Según la relación entre $\alpha^2$ y $\omega_0^2$, se distinguen tres casos:

 * Subamortiguado **($\alpha^2 <\omega_0^2$**) o ($b<2\sqrt{mk}$): Poco amortiguamiento y el sistema realiza varias oscilaciones antes de detenerse en la practica. La solución es de la forma:

 $x(t) = Ae^{-αt}\cos(w_0t+ϕ)$   $\quad$ (4)

* Criticamente amortiguado (**$\alpha^2 = \omega_0^2$**) o ($b=2\sqrt{mk}$): Suficiente amortiguamiento para que el sistema no llegue a oscilar y converja al origen tan rapido como sea posible. La solución es de la forma:

  $x(t) = C_1e^{-αt} + C_2te^{-αt}$  $\quad$ (5)

* Sobreamortiguado (**$\alpha^2 > \omega_0^2$)** o ($b>2\sqrt{mk}$): Demasiado amortiguamiento para que el sistema no llegue a oscilar y converja al origen de forma más lenta. La solución es de la forma:

  $x(t) = C_1e^{\lambda_1 t} + C_2e^{\lambda_2 t}$   $\quad$ (6)

  donde:

  $\lambda_1 = -α+\sqrt{\alpha^2 - \omega_0^2}$ $\quad$ (7)

  $\lambda_2 =-α-\sqrt{\alpha^2 - \omega_0^2}$ $\quad$ (8)


En todos los casos las constantes se calculan con las condiciones iniciales.

**Solución particular**

Para cada uno de los tres casos se calculan las constantes de integración usando las condiciones iniciales tanto para la osición $x$ como para la velocidad $v$. Por tal razón, para cada caso debemos reemplazar las condiciones iniciales en la solucion general para cada uno de los tres casos.  

 * **Caso Sobreamortiguado**

 $x(t) = Ae^{-αt}\cos(\omega_0t+ϕ)$    $\qquad$ (9)

 $v(t)=\frac{dx}{dt} = -A\omega e^{-αt}\sin(\omega t+ϕ) - A\alpha e^{-αt} \cos(\omega t+ϕ)$.  $\qquad$ (10)

 donde: $\omega = \sqrt{\omega_0^2-\alpha^2}$  $\qquad$ 11)

  Reemplazando las condiciones iniciales dads por (2) tenemos:

  $x_0 = A\cos(ϕ)$    $\qquad$ (12)

  $v_0 = -A\omega \sin(ϕ) - A\alpha \cos(ϕ)$.  $\qquad$ (13)

  Resolviendo el sistema (12) y (13):

  $A = \sqrt{x_0^2 + (\frac{v_0 + \alpha x_0}{\omega^2})^2}$ $\quad$ (14)

  $\alpha = \arctan(-\frac{v_0 + \alpha x_0}{\omega x_0})$  $\quad$ (15)



* **Caso critico**

 $x(t) = C_1e^{-αt} + C_2te^{-αt}$    $\qquad$ (16)

  $v(t)=\frac{dx}{dt} = -C_1\alpha e^{-αt} -C_2\alpha t e^{-αt} + C_2 e^{-αt}$ .  $\qquad$ (17)

 Reemplazando las condiciones iniciales dads por (2) tenemos:

  $x_0 = C_1$    $\qquad$ (18)

  $v_0 = -C_1\alpha$.  $\qquad$ (19)

  Resolviendo el sistema (18) y (19):

  $C_1 = x_0$  $\quad$ (20)

  $C_2 = v_0 + \alpha x_0$  $\quad$ (21)


* **Caso sobreamortiguado**

 $x(t) = C_1e^{\lambda_1 t} + C_2e^{\lambda_2 t}$    $\qquad$ (22)

  $v(t)=\frac{dx}{dt} = C_1 \lambda_1 e^{\lambda_1 t} + C_2 \lambda_2 e^{\lambda_2 t}$ .  $\qquad$ (23)

 Reemplazando las condiciones iniciales dads por (2) tenemos:

  $x_0 = C_1 + C_2$    $\qquad$ (24)

  $v_0 = C_1 \lambda_1 + C_2 \lambda_2$.  $\qquad$ (25)

  Resolviendo el sistema (9) y (10):

  $C_1 = \frac{v_0 - \lambda_2 x_0}{\lambda_1-\lambda_2}$ $\quad$ (26)

  $C_2 = -(\frac{v_0 - \lambda_1 x_0}{\lambda_1-\lambda_2})$  $\quad$ (27)


Las constantes calculadas se reemplazan en la solución general y se obtienen las soluciones particulares para una determinada posición y velocidad al inicio del movimiento.



# **Sistema masa-resorte amortiguado con Python**

Ahora, dado un valor de los parámetros del sistema ($k$, $m$ y $b$) y una posición y velocidad inicial ($x_0$ y $v_0$), obtendremos la solución particilar $x(t)$ y haremos su gráfica.

```python
#@title **(1) Subamortiguado: b^2 < 4*m*k**

import numpy as np
import matplotlib.pyplot as plt

m = 1.0    # Masa
k = 5.0    # Constante elástica
x0 = 1.0   # Posición inicial
v0 = 0.0   # Velocidad inicial
f = 0.2; # numero de veces b_critico
b_sub = f*(2*np.sqrt(m*k)) # coeficiente de amortiguamiento

Alpha = b_sub/(2*m)
Omega_0 = np.sqrt(k/m) # frecuencia natural (sin amortiguamiento)
Omega= np.sqrt(Omega_0**2 - Alpha**2) # frecuencia sistema amortiguado

# Tiempo de simulación
t = np.linspace(0, 10, 300)
# constantes de integracion
A = np.sqrt(x0**2 + (v0 + Alpha*x0)/(Omega**2))
Phi = np.arctan(-(v0+Alpha*x0)/(Omega*x0))
# Solucion paticular
x_sub = A*np.exp(-Alpha*t)*np.cos(Omega*t + Phi)

# grafica de x
plt.figure(figsize=(8,6))
plt.plot(t, x_sub, label='Subamortiguado (b = {})'.format(b_sub))
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.title('Oscilador Armónico Amortiguado')
plt.legend()
plt.grid(True)
plt.show()

print('Comparacion de las frecuencias angulares:')
print("Frecuencia angular (sin amortiguamiento):", Omega_0)
print("Frecuencia angular (con amortiguamiento):", Omega)
```
